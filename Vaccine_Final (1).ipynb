```python
#Import all packages needed

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
from datetime import datetime
from dateutil.parser import parse
```

```python
#import vaccine data
df1 = pd.read_csv('/Users/ken/Documents/random_projects/country_vaccinations-2.csv', parse_dates = [3])
```

```python
#Oh look the 'date' column is a string
type(df1['date'])
```

```python
#Change the date column form the first dataframe to date format

df1['date'] = pd.to_datetime(df1['date'])
```

```python
#Verify that the date column is in Timestamp format

type(df1['date'][1])
```

```python
#import second dataset: the Covid cases dataset

df2 = pd.read_csv('/Users/ken/Documents/random_projects/countries-aggregated.csv', parse_dates = True)
```

```python
#change the name of the "Date" column to "date" to match the vaccine database

df2.columns = ['date', 'country' , 'Confirmed',  'Recovered',  'Deaths']
print(df2)
```

```python
#Change the date column form the second dataframe to date format

df2['date'] = pd.to_datetime(df2['date'])
```

```python
#Verify that the date column is in Timestamp format

type(df2['date'][1])
```

```python
#Get rid of all dates before 12/13 -- when vaccine data starts

df2 = df2[~(df2['date'] < '12-13-2020')]

```

```python
#merge the two dataframes on the date column

df_final = df1.merge(df2, on=['country', 'date'])
```

```python
#Create a new date column called date_string that is the date column converted into the appropriate string format which will allow us to annimate the graph

df_final['date_string'] = df_final['date'].dt.strftime('%m/%d/%Y')
```

```python
#For the 3rd and 16th columns we have to replace any Nan values with the values from the previous day to make sure we can get a continous story from the data. 
#This data set is not perfect so this is a good fix to work around any missing values without misrepresenting the data

df_final = df_final.replace(3, np.nan).ffill()
df_final = df_final.replace(16, np.nan).ffill()
```

```python
#Similar to the line above but replaces all nan values with 0

df_final = df_final.replace(np.nan, 0) 
```

```python
##THIS IS THE ISSUE. As of right now I cannot correctly order the dates that control the animation frames

df_final = df_final.sort_values(['date_string'], ascending=[True])
```

```python
print(df_final)
```

```python
#Create the graph 

import plotly.express as px

fig = px.scatter(df_final, x="total_vaccinations", 
                 y="Deaths", 
                 size="Confirmed", 
                 animation_frame="date_string",
                 color='vaccines',
                 animation_group="country", 
                 hover_name="country", 
                 width=1400, 
                 height=800,

 log_x=True, size_max=200, range_x=[100,12000000], range_y=[1,700000])
fig.update_layout(
 title='Deaths vs Vaccines',
 yaxis=dict(
 title='Number of Deaths',
 gridcolor='white',
 gridwidth=2,
 ))
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
for i in range(1,7):
    fig.update_xaxes(title_text="Number of Vaccines Given", row=1, col=i)
fig.show()
```
